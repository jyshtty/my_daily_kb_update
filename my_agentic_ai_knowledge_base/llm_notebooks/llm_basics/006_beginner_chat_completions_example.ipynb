## Chat Completions API Demo

Demonstrates how a multi-turn dialog is composed and how the OpenAI API is used to call GPT and get a reply.

````python
%pip install -q openai python-dotenv
````

````python
import os
from dotenv import load_dotenv

load_dotenv()
assert os.environ.get("OPENAI_API_KEY"), "Set OPENAI_API_KEY in a .env file"
print("API key loaded.")
````

### Option A — Standard OpenAI

````python
from openai import OpenAI

client = OpenAI()  # reads OPENAI_API_KEY from env
model = "gpt-4o-mini"
````

### Option B — EPAM DIAL (requires EPAM VPN + DIAL key set as OPENAI_API_KEY)

```python
from openai import AzureOpenAI
client = AzureOpenAI(
    api_key=os.environ["OPENAI_API_KEY"],
    api_version="2023-07-01-preview",
    azure_endpoint="https://ai-proxy.lab.epam.com",
)
model = "gpt-4o-mini-2024-07-18"
```

### Build and extend a multi-turn conversation

````python
# Recreate an existing conversation
message_text = [
    {"role": "system",    "content": "You are an AI assistant that helps people find information."},
    {"role": "user",      "content": "Hello"},
    {"role": "assistant", "content": "Hello! How can I assist you today?"},
    {"role": "user",      "content": "What is the distance between Earth and Moon?"},
    {"role": "assistant", "content": "The average distance between the Earth and the Moon is approximately 384,400 kilometres."},
]

# Append the next user message
message_text.append({"role": "user", "content": "What is the distance to Mars?"})

# Call the API
completion = client.chat.completions.create(
    model=model,
    messages=message_text,
    temperature=0.7,
    max_tokens=800,
)

reply = completion.choices[0].message
print(reply.role + ": " + reply.content)
````

### Key points

- The full conversation history is sent on every request — the model has no persistent memory of its own.
- `role` can be `system`, `user`, or `assistant`.
- `temperature` controls randomness (0 = deterministic, 1 = creative).
- `max_tokens` caps the response length.
